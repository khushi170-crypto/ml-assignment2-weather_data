```python
import requests
import pandas as pd
url = "https://api.open-meteo.com/v1/forecast?latitude=22.7179&longitude=75.8333&current_weather=true"
print("fetching weather data...........")
response = requests.get(url)
if response.status_code  == 200:
  raw_data = response.json()
  weather_info = raw_data['current_weather']
  formatted_data = [{
      "time":weather_info['time'],
      "temperature °c":weather_info['temperature'],
      "wind speed (km/h)":weather_info['windspeed'],
      "wind direction":weather_info['winddirection']

  }]
  df = pd.DataFrame(formatted_data)
  print("data formatted successfully")
  print(df)
  df.to_csv("weather_data.csv",index=False)
  print("csv file created and saved")
else:
  print(response.status_code)
```
